```python
file1 = open('logs.txt', 'r')
Lines = file1.readlines()
 
count = 0
# Strips the newline character
numbers = []
for line in Lines:
    if(line[0] != 'E'): 
        continue
    
    count += 1
    index = line.rfind("=") +1
    numbers.append(float(line[index:]))
```

```python
print(numbers)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(numbers)
plt.title("Total Rewards")
plt.show()
```
